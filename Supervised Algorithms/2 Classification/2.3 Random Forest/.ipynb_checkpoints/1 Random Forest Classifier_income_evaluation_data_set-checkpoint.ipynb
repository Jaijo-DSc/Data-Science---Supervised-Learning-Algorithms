```python
df.drop(["relationship","education","fnlwgt","native-country"],axis =1, inplace =True)
```

```python
# encoding the categorical variable
df =pd.get_dummies(df)
```

```python
df
```

```python
df.shape
```

```python
df['income_1'] = df['income_ <=50K']
df['income_2'] = df['income_ >50K']
```

```python
df.drop(['income_ <=50K','income_ >50K'],axis =1, inplace =True)
```

```python
df.head()
```

```python
#splitting the dependent and independent variable
x = df.drop(['income_1', 'income_2'],axis=1)
y=df.income_1
```

```python
x.columns
```

```python
#splitting into training and testing dataset
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y)
```

```python
forest =RandomForestClassifier(n_estimators=100, max_features=0.4,max_depth= 20, random_state=121)
forest.fit(x_train,y_train)
```

```python
forest.score(x_train,y_train)
```

```python
forest.score(x_test,y_test)
```

```python
importances = forest.feature_importances_
```

```python
# Sort feature importances in descending order
indices = np.argsort(importances)[::-1]

# Rearrange feature names so they match the sorted feature importances
names = [x_train.columns[i] for i in indices]

# Create plot
plt.figure(figsize =(20,5))

# Create plot title
plt.title("Feature Importance")

# Add bars
plt.bar(range(x_train.shape[1]), importances[indices])

# Add feature names as x-axis labels
plt.xticks(range(x_train.shape[1]), names, rotation=90)

# Show plot
plt.show()
```
