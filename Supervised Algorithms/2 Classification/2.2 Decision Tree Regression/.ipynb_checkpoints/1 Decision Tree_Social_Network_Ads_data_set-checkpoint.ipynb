```python
#imputation
#"dropping" the columns which are not necessary
df.drop("User ID",axis = 1,inplace =True)
```

```python
#checking if any missing values are there
df.isnull().sum()
```

```python
# we have to convert Gender in such a form that alogorith can work on
df =pd.get_dummies(df)
```

```python
df.head()
```

```python
#splitting the dependent and independent variable
x = df.drop(['Purchased'], axis =1)
y = df.Purchased
```

```python
#splitting the training and testing dataset
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=101)
```

# model training
dtc = DecisionTreeClassifier(criterion='gini', max_depth=10,min_samples_leaf=5)
dtc.fit(x_train,y_train)

```python
# model training (if entropy used as criteria)
dtc = DecisionTreeClassifier(criterion="entropy", max_depth=10,min_samples_leaf=5)
dtc.fit(x_train,y_train)
```

```python
y_predict  = dtc.predict(x_test)
```

```python
accuracy_score(y_predict,y_test)
```

```python
confusion_matrix(y_predict,y_test)
```

```python
from sklearn.tree import export_graphviz
from sklearn.externals.six import StringIO
from IPython.display import Image
import pydotplus
```

```python
dot_data = StringIO()
export_graphviz(dtc,
                out_file=dot_data,
                filled=True,
                rounded=True,
                special_characters=True,
                feature_names=['Age','EstimatedSalary','Gender_Male','Gender_Female'],
                class_names=['0','1']
               )
graph=pydotplus.graph_from_dot_data(dot_data.getvalue())
graph.write_png('ads1.png')
Image(graph.create_png())
```
